```python
import pandas as pd
import json
import os
```

```python
CSV_PATH = "csv/github_agent_repos_python_20251225.csv"

# Load CSV into a DataFrame
df = pd.read_csv(CSV_PATH)
```

```python
print(f"Total rows loaded: {len(df)}")
print(df.columns)  # show all column names
df.head(2) 
```

```python
FEW_SHOT_JSON = "data/few_shot_examples_1.json"


# Initialize empty examples list
examples = []
```

```python
if os.path.exists(FEW_SHOT_JSON):
    # Check if file is non-empty
    if os.path.getsize(FEW_SHOT_JSON) > 0:
        try:
            with open(FEW_SHOT_JSON, "r") as f:
                examples = json.load(f)
        except json.JSONDecodeError:
            print(f"Warning: {FEW_SHOT_JSON} is not valid JSON. Starting with empty list.")
            examples = []
    else:
        print(f"{FEW_SHOT_JSON} is empty. Starting with empty list.")
else:
    print(f"{FEW_SHOT_JSON} does not exist. Starting with empty list.")

print(f"Loaded {len(examples)} few-shot examples.")
```
