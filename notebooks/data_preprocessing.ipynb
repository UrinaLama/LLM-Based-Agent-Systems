```python
from langdetect import detect
import pandas as pd
```

```python
def detect_lang(text): 
    if not isinstance(text, str) or len(text.strip()) == 0:
        return "nan" 
    try: 
        return detect(text) 
    except:
        return "unknown"
```

```python
df_sample = pd.read_csv("csv/sample_for_llm_shortReadme.csv")
len(df_sample)
```

```python
null_readme = df_sample["readme_snippet"].isna() | (df_sample["readme_snippet"].str.strip() == "")
print(f"Number of null or empty readme_snippet: {null_readme.sum()}")
df_sample[null_readme].head(3)
```

```python
df_sample = df_sample.drop(df_sample[null_readme].index)
len(df_sample)
```

```python
df_sample["readme_lang"] =  df_sample["readme_snippet"].apply(detect_lang) 
df_sample["desc_lang"] = df_sample["description"].apply(detect_lang)
```

```python
df_notenglish = df_sample[(df_sample["readme_lang"] != "en") & (df_sample["desc_lang"] != "en")]
print("Number of repositories not in English:", len(df_notenglish))
df_notenglish
```

```python
df_clean = df_sample.drop(df_notenglish.index)
df_clean = df_clean.drop(columns=["desc_lang", "readme_lang"])
```

```python
print("Length of cleaned dataset:", len(df_clean))
df_clean.head(3)
```

```python
df_clean.to_csv("csv/clean_sample_for_llm_shortReadme.csv", index=False)
```
