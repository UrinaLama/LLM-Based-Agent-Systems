```python
def build_few_shot(path="data/few_shot_examples.json", max_readme_words=3000):
    with open(path, "r") as f:
        examples = json.load(f)

    blocks = []

    for i, ex in enumerate(examples, 1):
        #readme = ex["readme"][:max_readme_chars]
        # Prefer readme_content, fallback to readme_snippet
        readme = ex['readme_content'] 

        if not isinstance(readme, str):
            readme = ""

        words = readme.split()
        if len(words) > max_readme_words:
            readme = " ".join(words[:max_readme_words]) + "\n\n[TRUNCATED]"

        blocks.append(f"""
        ### Example {i}
        Input:
        name: "{ex['full_name']}"
        description: "{ex['description']}"
        topics: {ex['topics']}
        readme:
        \"\"\"
        {readme}
        \"\"\"

        Output:
        {{"category":"{ex['category']}"}}
        """)

    return "\n".join(blocks)
```

```python
FEW_SHOT = build_few_shot(FEW_SHOT_JSON, max_readme_words=3000)
print(FEW_SHOT)
```
