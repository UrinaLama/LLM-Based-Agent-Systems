```python
from github import Github, Auth

# GitHub Personal Access Token
GITHUB_TOKEN = ""

# Use new Auth method
auth = Auth.Token(GITHUB_TOKEN)

# Initialize GitHub client with the new auth object
g = Github(auth=auth)
```

```python
# Search query
llm_keywords = ["LLM", '"large language model"']
agent_keywords = ["agent", "autonomous", '"multi-agent"']

filters = "language:Python fork:false archived:false created:>=2022-01-01"

all_repos = {}

for llm in llm_keywords:
    for agent in agent_keywords:
        query = f'{llm} {agent} {filters}'
        print("\nRunning query:", query)

        results = g.search_repositories(query=query)

        for repo in results:
            # Use repo.full_name as a unique ID — avoids duplicates
            all_repos[repo.full_name] = repo

print("\nTotal unique repositories found:", len(all_repos))

# Print repo list
for name, repo in all_repos.items():
    print(name, ", ⭐", repo.stargazers_count)
```

**Phase 1: Mining Repositories**

```python
import csv
```

```python
output_file = "github_repos.csv"

with open(output_file, mode="w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)

    # Header
    writer.writerow([
        "full_name",
        "description",
        "stars",
        "forks",
        "language",
        "url"
    ])

    # Rows
    for name, repo in all_repos.items():
        #topics = ", ".join(repo.get_topics())  # get topics as comma-separated string
        writer.writerow([
            repo.full_name,
            repo.description or "",
            repo.stargazers_count,
            repo.forks_count,
            repo.language or "",
            repo.html_url
        ])

print("CSV file created:", output_file)
```

**Phase 2: Filter out the benchmark, collection, environment, framework from phase 1**

```python
output_files = {
    "handbook": "handbook.csv",
    "framework": "framework_or_foundation.csv",
    "collection": "collection.csv",
    "evaluation": "evaluation_or_benchmark.csv",
    "agent_repos": "agent_repos.csv",
}

files = {key: open(fname, "w", newline="", encoding="utf-8")
         for key, fname in output_files.items()}

writers = {key: csv.writer(f) for key, f in files.items()}

header = [
    "full_name",
    "description",
    "stars",
    "forks",
    "language",
    "url",
    "year",
    "last_updated",
]

for w in writers.values():
    w.writerow(header)

#keywords to filter out
handbook = ["handbook"]
framework = ["framework", "foundation model"]
collection = ["collection", "curated"]
evaluation = [ "benchmark", "evaluate"]

def classify_repo(name, description):
    text = f"{name} {description}".lower()

    if any(k in text for k in handbook):
        return "handbook"
    if any(k in text for k in framework):
        return "framework"
    if any(k in text for k in collection):
        return "collection"
    if any(k in text for k in evaluation):
        return "evaluation"

    return "agent_repos"
```

```python
for name, repo in all_repos.items():

    desc = repo.description or ""
    year = repo.pushed_at.year if repo.pushed_at else ""
    last_updated = repo.pushed_at.isoformat() if repo.pushed_at else ""

    row = [
        repo.full_name,
        desc,
        repo.stargazers_count,
        repo.forks_count,
        repo.language or "",
        repo.html_url,
        year,
        last_updated
    ]

    category = classify_repo(repo.full_name, desc)
    writers[category].writerow(row)

for f in files.values():
    f.close()

print("All category CSV files created successfully!")
```

```python
def count_rows(filename):
    with open(filename, "r", encoding="utf-8") as f:
        return sum(1 for _ in f) - 1   # subtract header

for label, fname in output_files.items():
    print(f"{fname}: {count_rows(fname)} records")
```
